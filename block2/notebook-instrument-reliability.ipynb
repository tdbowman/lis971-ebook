## Cronbach's Alpha Calculator

Cronbach's alpha measures internal consistency — how well a set of items measures the same underlying construct. Use this tool with your pilot data to assess your instrument before full data collection.

```python
import numpy as np

def cronbachs_alpha(items):
    """
    Calculate Cronbach's alpha for a set of survey items.

    Parameters
    ----------
    items : array-like, shape (n_respondents, n_items)
        Matrix of item responses
    """
    items = np.array(items, dtype=float)
    n_items = items.shape[1]
    item_variances = items.var(axis=0, ddof=1)
    total_variance = items.sum(axis=1).var(ddof=1)
    alpha = (n_items / (n_items - 1)) * (1 - item_variances.sum() / total_variance)
    return alpha

# ─── EXAMPLE: Simulated 5-item Likert scale (30 respondents) ───
np.random.seed(42)
n_respondents, n_items = 30, 5
base = np.random.normal(3.5, 1.0, n_respondents)
items = np.column_stack([
    np.clip(base + np.random.normal(0, 0.5, n_respondents), 1, 5)
    for _ in range(n_items)
]).round()

alpha = cronbachs_alpha(items)

print("=" * 50)
print("  CRONBACH'S ALPHA CALCULATOR")
print("=" * 50)
print(f"  Respondents:  {n_respondents}")
print(f"  Items:        {n_items}")
print(f"  Alpha:        {alpha:.3f}")
print("─" * 50)
if alpha >= 0.90: interp = "Excellent"
elif alpha >= 0.80: interp = "Good"
elif alpha >= 0.70: interp = "Acceptable"
elif alpha >= 0.60: interp = "Questionable"
else: interp = "Poor — consider revising items"
print(f"  Interpretation: {interp}")
print(f"\n  Convention: α ≥ 0.70 is generally acceptable")
print(f"  (Nunnally & Bernstein, 1994)")
print("=" * 50)
print("\n>>> Replace the simulated data with your own pilot data!")
```

## Item-Level Analysis

This tool shows what happens to alpha if each item is removed — useful for identifying problematic items that reduce internal consistency.

```python
def item_analysis(items, item_labels=None):
    """Show alpha-if-item-deleted for each item."""
    items = np.array(items, dtype=float)
    n_items = items.shape[1]
    if item_labels is None:
        item_labels = [f"Item {i+1}" for i in range(n_items)]
    
    full_alpha = cronbachs_alpha(items)
    print(f"Overall Cronbach's alpha: {full_alpha:.3f}\n")
    print(f"{'Item':<15} {'α if deleted':<15} {'Action'}")
    print("─" * 45)
    
    for i in range(n_items):
        reduced = np.delete(items, i, axis=1)
        reduced_alpha = cronbachs_alpha(reduced)
        flag = "← CONSIDER REMOVING" if reduced_alpha > full_alpha + 0.01 else ""
        print(f"{item_labels[i]:<15} {reduced_alpha:.3f}{'':>10} {flag}")

item_analysis(items, ["Q1_Access", "Q2_Quality", "Q3_Ease", "Q4_Speed", "Q5_Satisfy"])
```
