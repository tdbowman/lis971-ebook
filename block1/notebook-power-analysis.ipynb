## Power Analysis Calculator (Quantitative)

This tool calculates required sample sizes for common statistical tests using the formulas from Cohen (1992).

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def power_analysis(test_type="t-test", effect_size="medium",
                   alpha=0.05, power=0.80):
    """
    Calculate required sample size using Cohen's (1992) conventions.

    Parameters
    ----------
    test_type : str
        One of: 't-test', 'ANOVA', 'correlation', 'chi-square'
    effect_size : str or float
        'small', 'medium', 'large', or a numeric value
    alpha : float
        Significance level (default 0.05)
    power : float
        Statistical power (default 0.80)
    """
    conventions = {
        "t-test":      {"small": 0.20, "medium": 0.50, "large": 0.80},
        "ANOVA":       {"small": 0.10, "medium": 0.25, "large": 0.40},
        "correlation": {"small": 0.10, "medium": 0.30, "large": 0.50},
        "chi-square":  {"small": 0.10, "medium": 0.30, "large": 0.50}
    }

    if isinstance(effect_size, str):
        d = conventions[test_type][effect_size]
    else:
        d = effect_size

    z_alpha = stats.norm.ppf(1 - alpha/2)
    z_beta = stats.norm.ppf(power)

    if test_type == "t-test":
        n_per_group = math.ceil(((z_alpha + z_beta) / d) ** 2) + 1
        total_n = n_per_group * 2
        label = f"n per group = {n_per_group}, total N = {total_n}"
    elif test_type == "correlation":
        n = math.ceil(((z_alpha + z_beta) / (0.5 * math.log((1+d)/(1-d)))) ** 2) + 3
        total_n = n
        label = f"N = {total_n}"
    elif test_type == "ANOVA":
        n_per_group = math.ceil(((z_alpha + z_beta) / d) ** 2) + 1
        total_n = n_per_group
        label = f"n per group ≈ {n_per_group} (multiply by number of groups)"
    else:
        n = math.ceil(((z_alpha + z_beta) / d) ** 2)
        total_n = n
        label = f"N ≈ {total_n}"

    print(f"{'='*55}")
    print(f"  POWER ANALYSIS RESULTS")
    print(f"{'='*55}")
    print(f"  Test:         {test_type}")
    print(f"  Effect size:  {effect_size} (d/f/r = {d})")
    print(f"  Alpha:        {alpha}")
    print(f"  Power:        {power}")
    print(f"  ───────────────────────────────────")
    print(f"  Required:     {label}")
    print(f"{'='*55}")
    print(f"\n  Citation: Cohen, J. (1992). A power primer.")
    print(f"  Psychological Bulletin, 112(1), 155-159.")
    return total_n

# ─── MODIFY THESE PARAMETERS FOR YOUR STUDY ───
n = power_analysis(
    test_type="t-test",      # Options: t-test, ANOVA, correlation, chi-square
    effect_size="medium",    # Options: small, medium, large, or a number
    alpha=0.05,              # Significance level
    power=0.80               # Statistical power
)
```

## Power Curve Visualization

This plot shows how required sample size changes as a function of effect size and desired power. Use it to understand the trade-offs in your study design.

```python
def plot_power_curve(test_type="t-test", alpha=0.05):
    """Plot sample size as a function of effect size and power."""
    effect_sizes = np.linspace(0.15, 1.0, 50)
    powers = [0.70, 0.80, 0.90, 0.95]
    colors = ["#7b5ea7", "#2c5f7c", "#4a9068", "#c0612b"]

    fig, ax = plt.subplots(figsize=(10, 6))

    for pwr, color in zip(powers, colors):
        z_alpha = stats.norm.ppf(1 - alpha/2)
        z_beta = stats.norm.ppf(pwr)
        sample_sizes = [2 * (math.ceil(((z_alpha + z_beta) / d) ** 2) + 1)
                        for d in effect_sizes]
        ax.plot(effect_sizes, sample_sizes, color=color, linewidth=2,
                label=f"Power = {pwr}")

    ax.set_xlabel("Effect Size (Cohen's d)", fontsize=11)
    ax.set_ylabel("Total Sample Size (N)", fontsize=11)
    ax.set_title(f"Required Sample Size by Effect Size and Power\n"
                 f"({test_type}, α = {alpha}, two-tailed)",
                 fontsize=12, fontweight="bold")
    ax.legend(fontsize=10)
    ax.set_ylim(0, 1000)
    ax.grid(True, alpha=0.3)

    # Cohen's conventions
    for label, d in {"Small\n(0.20)": 0.20, "Medium\n(0.50)": 0.50, "Large\n(0.80)": 0.80}.items():
        ax.axvline(x=d, color="gray", linestyle="--", alpha=0.5)
        ax.text(d, ax.get_ylim()[1]*0.95, label, ha="center", fontsize=8, color="gray")

    plt.tight_layout()
    plt.show()

plot_power_curve()
```

## Worldview Comparison Chart

This visualization maps the four major worldviews across philosophical dimensions. Use it to identify where your own assumptions cluster.

```python
categories = ["Ontology\n(Nature of Reality)",
              "Epistemology\n(Knower–Known)",
              "Axiology\n(Role of Values)",
              "Methodology\n(Research Process)",
              "Rhetoric\n(Language of Research)"]

worldviews = {
    "Postpositivism": ["Single reality\n(apprehendable)", "Objectivist\n(detached observer)",
                       "Value-free\n(bias controlled)", "Deductive\n(hypothesis testing)",
                       "Formal\n(impersonal voice)"],
    "Constructivism": ["Multiple realities\n(co-constructed)", "Subjectivist\n(co-created meaning)",
                        "Value-laden\n(acknowledged)", "Inductive\n(emergent design)",
                        "Informal\n(personal voice)"],
    "Transformative": ["Political reality\n(shaped by power)", "Collaborative\n(with participants)",
                        "Value-driven\n(social justice)", "Participatory\n(action-oriented)",
                        "Advocacy\n(change language)"],
    "Pragmatism": ["Practical reality\n(what works)", "Pluralistic\n(multiple stances)",
                    "Value-aware\n(problem-centered)", "Abductive\n(mixed approaches)",
                    "Flexible\n(matched to purpose)"]
}

fig, ax = plt.subplots(figsize=(14, 7))
colors = ["#2c5f7c", "#4a9068", "#c0612b", "#7b5ea7"]
y_positions = np.arange(len(categories))
bar_height = 0.18

for i, (wv, descriptions) in enumerate(worldviews.items()):
    offset = (i - 1.5) * bar_height
    ax.barh(y_positions + offset, [1]*len(categories),
            height=bar_height, color=colors[i], alpha=0.85, label=wv)
    for j, desc in enumerate(descriptions):
        ax.text(0.5, y_positions[j] + offset, desc,
                ha="center", va="center", fontsize=7.5, color="white", fontweight="bold")

ax.set_yticks(y_positions)
ax.set_yticklabels(categories, fontsize=10, fontweight="bold")
ax.set_xlim(0, 1)
ax.set_xticks([])
ax.legend(loc="upper right", fontsize=9, framealpha=0.9)
ax.set_title("Four Worldviews Across Philosophical Dimensions\n(Adapted from Creswell & Creswell, 2023; Crotty, 1998)",
             fontsize=12, fontweight="bold", pad=15)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
plt.tight_layout()
plt.show()
```
